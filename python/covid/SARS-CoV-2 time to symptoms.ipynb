Based on [this paper](https://bmjopen.bmj.com/content/bmjopen/10/8/e039652.full.pdf), which is a meta-analysis of the incubation period for SARS-CoV-2.



They approximate it as a lognormal distribution, with $\mu=1.63$ and $\sigma=0.50$.

```python
# Based on https://stackoverflow.com/a/13837335
import numpy as np
import pylab as plt
from scipy.stats import lognorm
from matplotlib.ticker import MaxNLocator

σ = 0.50
μ = 1.63
dist=lognorm(s=σ, scale=np.exp(μ))

x=np.linspace(0, 20, 200)
fig = plt.figure(figsize=(12,8), dpi= 100, facecolor='w', edgecolor='k')
ax = fig.gca()

ticks = np.arange(0, 21, 1.0)
plt.vlines(ticks, 0, dist.cdf(ticks), linestyle="dashed")
# plt.hlines(dist.cdf(ticks), 0, ticks, linestyle="dashed")
# plt.scatter(x, y, zorder=2)
for tick in ticks:
    cdf = dist.cdf(tick)
    plt.text(tick, dist.cdf(tick), f"{cdf:.2f}" ,horizontalalignment='right')
plt.scatter(ticks, dist.cdf(ticks))

# plt.plot(x, dist.pdf(x))
plt.plot(x, dist.cdf(x))
# ax.xaxis.set_major_locator(MaxNLocator(integer=True))
_ = plt.xticks(ticks)
```
