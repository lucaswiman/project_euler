### Notes:

"Markov with respect to" is defined on page 16, definition 1.2.2. A probability function $P$ is _Markov_ w.r.t. a graph $G$ iff:
$$P(x_1, \ldots, x_n) = \prod_i P(x_i|pa_i)$$
where $pa_i$ denotes the _parents_ of $x_i$.

"Sets are anything that doesn't have a subscript $i$" - @rxia



Proof:

$\Leftarrow$ Suppose that the truncated factorization holds for all interventions $X=x$ and for all $v$ consistent with $x$. Then:

(i) Let $X=\emptyset$. Then the truncated factorization implies that:
$$ P_x(v) = \prod_{\{i|V_i\not\in \emptyset\}} P\left(v_i|pa_i\right) $$
This is just the markov condition.

(ii) Expand using (i), and cancel like terms.

(iii) Write out $P_x(V=v_i)$ in two ways: using 1.37 and also using the markov property. Then both of those products have exactly one term in them, which must therefore be equal to each other.

$\Rightarrow$ Now suppose that (i)-(iii) hold. Then you can write out a product using (i), pull out terms that are 1 using (ii), and then drop the $x$ 

## Question:

What does "$v$ is consistent with $x$" mean here?

$X=x$ is like an associative array:
$$\{\cdots,X_i: x_i, \cdots\} $$
Where $X_i\in X$. $v$ refers to a particular assignment $V=v$. So consistent with $X=x$, means that $V\subseteq X$, i.e. every assignment in $V=v$ is also in $X=x$.
