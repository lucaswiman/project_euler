```python
import pygraphviz as pgv
from IPython.display import Image, SVG, display
import networkx as nx

def draw(graph):
    svg = nx.nx_agraph.to_agraph(graph).draw(prog='dot',format='svg')
    display(SVG(svg))
```

```python
import networkx as nx
G = nx.DiGraph()
G.add_edge('Personal characterisitics', 'Choice')
G.add_edge('Personal characterisitics', 'W_0')
G.add_edge('Personal characterisitics', 'W_F')
G.add_edge('Choice', 'W_F')
G.add_edge('W_0', 'W_F')
draw(G)
```

```python
import networkx as nx
G = nx.DiGraph()
G.add_edge('Ward', 'Lollipop')
G.add_edge('Ward', 'Treatment')
G.add_edge('Treatment', 'Recovery')
G.add_edge('Depression', 'Recovery')
G.add_edge('Depression', 'Lollipop')
draw(G)
```
