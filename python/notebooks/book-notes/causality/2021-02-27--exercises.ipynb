Now we need to compute the regression coefficients given the structural equations:

\begin{aligned}
X &= U_1 \\
Z &= aX + U_2 \\

Now we need to compute the regression coefficients. Per eqn 1.22 (page 22), the regression coefficient of $Y$ with respect to $X$ can be computed as:
$$ R_{YX} = {\sigma_{XY} \over \sigma_X^2 } $$

Where $\sigma_{XY} = \cov(X, Y)$ is the covariance function, which satisfies the follwoing bilinearity properties: https://en.wikipedia.org/wiki/Covariance#Properties


Note that $aX$ is normally distributed with a mean of $a\cdot 0=0$ and $\sigma_{aX} = |a|\sigma_X = |a|$ (see operations (here)[https://en.wikipedia.org/wiki/Normal_distribution#Operations_on_a_single_normal_variable]). Recall that adding two normally distributed normal variables, we sum the means, and the sum the variances (square of standard deviations). Thus $Z$ is normally distributed with $\cov(Z, Z) = \sigma_Z^2 = 1 + a^2$

\begin{aligned}
R_{XZ} &= \frac{\cov(Z, X)}{\sigma_Z^2 } \\
       &= \frac{\cov(aU_1 + U_2, U_1)}{1+a^2 } \\
       &= \frac{a\cov(U_1, U_1) + a \cov(U_2, U_2)}{1 + a^2}\\
       &= \frac{a}{1 + a^2}
\end{aligned}

To compute $R_{ZY}$, we can treat $Z$ as normally distributed (sum of two normals is normal), and we end up with:
\begin{aligned}
R_{YZ} &= \frac{\cov(bZ, Z)}{\cov(Z, Z)} \\
       &= \frac{b\cov(Z, Z)}{\cov(Z, Z)} \\
       &= b
\end{aligned}

Note that by the linearity of expectation, we have $E[Z]=E[aX + U_2]=aE[X] + E[U_2]=aE[U_1]=a\cdot 0 + 0= 0$. A similar computation yields that $E[Y]=0$.

$$E[X|Z=z] = E[x] + R_{XZ}(z-E[Z]) = 0 + \frac{a}{1 + a^2}(z-0)= \frac{a z}{1 + a^2}$$

Now we need to compute $E[Y|Z=z]$ to use theorem 4.3.2. Since $R_{YZ}=b$:
$$E[Y|Z=z] = E[y] + R_{YZ}(z-E[Z]) = 0 + b(z) = bz$$
(This also follows from more straightforward arguments.)
