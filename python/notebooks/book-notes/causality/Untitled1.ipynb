```python
import numpy as np
n_samples = 100000000
U_x = np.random.normal(0, 1, n_samples)
X = U_x
U_y = np.random.normal(0, 1, n_samples)
Y = X / 3 + U_y
U_z = np.random.normal(0, 1, n_samples)
Z = Y / 16 + U_z
```

```python
tol = 1e-2
sol = Y[(np.abs(X - 1) < tol) & (np.abs(Z-3) < tol)]
print(sol.mean())
print(sol.shape)
```
